```python
from urllib.request import urlopen
import requests
from bs4 import BeautifulSoup as soup
import cv2
import matplotlib.pyplot as plt
```

```python
# show_image(book_frame.iloc[0]['image_link'])
```

```python
def show_image(dta):
    with open('inspiration.jpj', 'wb') as file:
        url = requests.get(dta)
        file.write(url.content)
        img = cv2.imread('/content/inspiration.jpj')
        # rgb_image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        plt.style.use('seaborn')
# plt.imshow(img)
        plt.imshow(img)
        plt.axis("off")
        plt.show()
```

```python
def scrape(url, ans) :
    url_open = urlopen(url)
    book_html = url_open.read()
    url_open.close()
    url_soup = soup(book_html, "html.parser")
    # print(type(url_soup))
    a = url_soup.findAll('li')
    j = 0
    # k = 0
    book_data = []
    for i in a:
        j = j + 1
        if(j > 53):
            book_data.append(i)
    cnt = 0
    for data_book in book_data:  
        cnt = cnt+1
        if cnt>20:
            break
        data_book = (data_book.findAll('article'))
        lst = []
        # print('iuyefiuyiqwuyruwqyiurywiuyriwqyriuywiqryiwqy')
        # print(i)
        data_book = data_book[0].findAll('div')
    # for j in k:
    #     lst.append(j)
    # print(data_book)
        lst1 = []
        dta = data_book[0].findAll('a')[0].findAll('img')[0]
        name = dta.attrs['alt']
        img = "https://books.toscrape.com/"+dta.attrs['src']
        # show_image(img)
        stock_check = data_book[1].findAll('p')[1].text
        output = stock_check.split('\n')
        stock_check = output[3].lstrip()
        price = data_book[1].findAll('p')[0].text[1:]
        lst1.append(name.replace(',', '-'))
        lst1.append(img.replace(',', '-'))
        lst1.append(stock_check.replace(',', '-'))
        lst1.append(price.replace(',', '-'))
        # print(lst1)
        ans.append(lst1)
        # print()
```

```python
ans = list()
for i in range (1, 51, 1) :
    url = "https://books.toscrape.com/catalogue/page-{}.html".format(i)
    scrape(url, ans)
# print(ans)
```

```python
import pandas as pd
book_frame = pd.DataFrame(ans, columns = ['Name', 'image_link', 'Stock check', 'Price in Euros'], index=None)
```
